## Mixed $\{\Sigma_l\}_{l=1}^{L}$ by DP

*python script: https://github.com/letsjdosth/bnp/blob/main/hw4.py* (please run without commenting out the $\Sigma_l$ mixing part.)


Now, the model is the second scenario in my note, with mixing both $\mu_l$ and $\Sigma_l$ at the same time.
$G_0$'s hyperparameters for Inv.Wishart part is $v_0=3000$, $\Lambda_0=I_4$.
The giant $v_0$ implies the prior is strong, hoping the realization of $\Sigma_l$ is near $I_4$ without exploding too much.
Other hyperparameters are the same as the main report.

I run 500 iterations.
Let me skip reporting traceplots.

Joint posterior samples cover the area with observations, but the cover is too broad to explain the data points concisely.
Still, my setting suffers from inflated covariance.
Inverse Wishart distribution cannot generate a suitable covariance matrix for each $l$ in short iterations. 
And generally, its diagonal variance terms get so huge that it groups all data into one density component.
